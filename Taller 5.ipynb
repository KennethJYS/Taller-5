```python
from src import gauss_jacobi, gauss_seidel
import numpy as np
import matplotlib.pyplot as plt
```

#### Ejercicio 1

```python
A=[(2,10),(3,2)]
b=[16,11]
n=len(A)
x, tray=gauss_jacobi(A=A,b=b,x0=[1,1],tol=0.01, max_iter=100)
print("Soluci贸n aproximada:", x.T)
print("Iteraciones realizadas:", len(tray))
```

```python
plt.figure(figsize=(10, 5))
tray_array = np.array([t.flatten() for t in tray])
plt.plot(tray_array[:, 0], tray_array[:, 1], marker='o', linestyle='--', color='red')
plt.title('Gauss Jacobi')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```

```python
A=[(2,10),(3,2)]
b=[16,11]
n=len(A)
x, tray=gauss_seidel(A=A,b=b,x0=[1]*n,tol=1e-1, max_iter=100)
print("Soluci贸n aproximada:", x.T)
print("Iteraciones realizadas:", len(tray))
```

```python
plt.figure(figsize=(10, 5))
tray_array = np.array([t.flatten() for t in tray])
plt.plot(tray_array[:, 0], tray_array[:, 1], marker='o', linestyle='--', color='blue')
plt.title('Gauss seidel')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```

## Ejercicio 2

```python
A=[(3,2),(2,10)]
b=[11,16]
n=len(A)
x, tray =gauss_jacobi(A=A,b=b,x0=[1,1],tol=0.001, max_iter=100)
print("Soluci贸n aproximada:", x.T)
print("Iteraciones realizadas:", len(tray))
```

```python
A=[(3,2),(2,10)]
b=[11,16]
n=len(A)
x, tray2 =gauss_seidel(A=A,b=b,x0=[1,1],tol=0.001, max_iter=100)
print("Soluci贸n aproximada:", x.T)
print("Iteraciones realizadas:", len(tray2))
```
